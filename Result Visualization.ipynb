```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as mpatches
```

```python
smote_thresh_tuned = pd.read_csv('hasil threshold smote.csv')
smote_thresh_base = pd.read_csv('hasil baseline smote threshold.csv')
rus_thresh_tuned = pd.read_csv('hasil threshold undersampled.csv')
rus_thresh_base = pd.read_csv('hasil baseline rus threshold.csv')

smote_thresh_base.drop(['Unnamed: 0'], axis=1, inplace=True)
smote_thresh_tuned.drop(['Unnamed: 0'], axis=1, inplace=True)
rus_thresh_base.drop(['Unnamed: 0'], axis=1, inplace=True)
rus_thresh_tuned.drop(['Unnamed: 0'], axis=1, inplace=True)

print(f'smote_thresh_base: {smote_thresh_base.shape}')
print(f'smote_thresh_tuned: {smote_thresh_tuned.shape}')
print(f'rus_thresh_base: {rus_thresh_base.shape}')
print(f'rus_thresh_tuned: {rus_thresh_tuned.shape}')
```

```python
smote_thresh_base.head()
```

## Recall

```python
# plot graph for recall value on smote base and smote thresh

# Create the line plots with markers
sns.lineplot(data=smote_thresh_base, x='Threshold', y='Recall', color='red', marker='o')
sns.lineplot(data=smote_thresh_tuned, x='Threshold', y='Recall', color='blue', marker='s')

# Set the labels and title
plt.xlabel('Threshold')
plt.ylabel('Recall')
plt.title('Recall on Baseline and Tuned Model using Oversampled Data')

# Create custom legend handles with desired colors and markers
red_patch = mpatches.Patch(color='red', label='Base Model')
blue_patch = mpatches.Patch(color='blue', label='Tuned Model')

# Create the legend with the custom handles
plt.legend(handles=[red_patch, blue_patch], title='Model', loc='upper right')

# Show the plot
plt.show()
```

```python
# plot graph for recall value on undersampled base and undersampled thresh

# Create the line plots with markers
sns.lineplot(data=rus_thresh_base, x='Threshold', y='Recall', color='red', marker='o')
sns.lineplot(data=rus_thresh_tuned, x='Threshold', y='Recall', color='blue', marker='s')

# Set the labels and title
plt.xlabel('Threshold')
plt.ylabel('Recall')
plt.title('Recall on Baseline and Tuned Model using Undersampled Data')

# Create custom legend handles with desired colors and markers
red_patch = mpatches.Patch(color='red', label='Base Model')
blue_patch = mpatches.Patch(color='blue', label='Tuned Model')

# Create the legend with the custom handles
plt.legend(handles=[red_patch, blue_patch], title='Model', loc='upper right')

# Show the plot
plt.show()
```
